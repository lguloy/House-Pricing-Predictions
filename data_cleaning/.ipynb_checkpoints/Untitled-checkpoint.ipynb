```python
Residential_df = train_df.loc[(train_df["MSZoning"]=="FV") | (train_df["MSZoning"]=="RH") | (train_df["MSZoning"]=="RL") | (train_df["MSZoning"]=="RP") | (train_df["MSZoning"]=="RM")]
```
